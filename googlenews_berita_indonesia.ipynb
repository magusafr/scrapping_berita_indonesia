## liputan6

```python
googlenews.search('liputan6 uu cipta kerja umkm')

df = pd.DataFrame()
for i in range(1,11):
  aaa = googlenews.page_at(i)
  df = df.append(aaa)

df['sumber'] = 'liputan6'
news_liputan6 = df.copy()
news_liputan6
```

# gabung

```python
df_semua = pd.concat([news_detik.reset_index().drop(columns=['index']), 
           news_tribun.reset_index().drop(columns=['index']), 
           news_pikiran.reset_index().drop(columns=['index']),
           news_merdeka.reset_index().drop(columns=['index']),   
           news_suara.reset_index().drop(columns=['index']), 
           news_liputan6.reset_index().drop(columns=['index']),
           df_lain
           ], ignore_index=True)

df_semua
```

```python
df_semua.to_csv("semua_link_news.csv", index=False)
```
